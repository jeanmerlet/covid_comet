```python
lineage_data = pd.read_csv(lineage_data_path, sep='\t')
lineage_data = lineage_data.sort_values(by='cluster_size', ascending=False)
lineage_data.index = list(range(lineage_data.shape[0]))
lineage_data = lineage_data.dropna()
```

```python
def cdc_strain_mapper(lineage):
    if lineage == 'Other':
        return 'None'
    elif re.match('B\.1\.1\.7', lineage) or re.match('Q', lineage):
        return 'Alpha'
    elif re.match('B\.1\.351', lineage):
        return 'Beta'
    elif re.match('P\.1', lineage):
        return 'Gamma'
    elif re.match('B\.1\.617\.2', lineage) or re.match('AY', lineage):
        return 'Delta'
    elif lineage == 'B.1.427' or lineage == 'B.1.429':
        return 'Epsilon'
    elif lineage == 'B.1.525':
        return 'Eta'
    elif lineage == 'B.1.526':
        return 'Iota'
    elif lineage == 'B.1.617.1':
        return 'Kappa'
    elif lineage == '1.617.3':
        return '1.617.3'
    elif lineage == 'P.2':
        return 'Zeta'
    elif lineage == 'B.1.621' or lineage == 'B.1.621.1':
        return 'Mu'
    elif lineage == 'B.1.1.529' or re.match('BA\.[12345]', lineage):
        return 'Omicron'
    else:
        return 'Other'
```

```python
all_lineages = []
for lineages in lineage_data.loc[:, 'top_lineages'].values:
    all_lineages.append(lineages.split(','))
all_lineages = [x for sublist in all_lineages for x in sublist]
all_lineages = list(np.unique(np.array(all_lineages)))
mapped_lineages = []
for lineage in all_lineages:
    mapped_lineages.append(cdc_strain_mapper(lineage))
mapped_lineages = list(np.unique(mapped_lineages))
num_lineages = len(mapped_lineages)
```

```python
# create new lineage size array based on renaming scheme

# get length of longest sizes
longest_size = 0
for i in range(lineage_data.shape[0]):
    current_size = len(lineage_data.loc[i, 'top_lineages'].split(','))
    if current_size > longest_size: longest_size = current_size
        
print(f'cluster with most lineages has: {longest_size}')

# first put unmapped sizes into array
unmapped_sizes = np.zeros((lineage_data.shape[0], longest_size))
for i, size_str in enumerate(lineage_data.loc[:, 'size_top_lineages'].values):
    split = size_str.split(',')
    unmapped_sizes[i, :len(split)] = split

# now map
col_names = mapped_lineages
index_names = np.array(list(range(lineage_data.shape[0])))
lineage_sizes = pd.DataFrame(data=np.zeros((lineage_data.shape[0], num_lineages)), index=index_names,
                             columns=col_names)
uniq_pango_lineages = []
for i, top_lineages_str in enumerate(lineage_data.loc[:, 'top_lineages']):
    lineage_names = top_lineages_str.split(',')
    uniq_pango_lineages.append(lineage_names) # just to count uniq pango lineages
    mapped_names = [cdc_strain_mapper(x) for x in lineage_names]
    for j, mapped_name in enumerate(mapped_names):
        lineage_sizes.loc[i, mapped_name] += unmapped_sizes[i, j]

uniq_pango_lineages = [x for sublist in uniq_pango_lineages for x in sublist]
print(f'number of unique pango lineages: {len(np.unique(uniq_pango_lineages))}')
```

```python
# specify n largest clusters
top_n = 100
# load cluster metadata file
clusters_meta = pd.read_csv(clusters_metadata_path, sep='\t')
clusters_meta = clusters_meta.sort_values(by='cluster_size', ascending=False).iloc[:top_n, :]

# sort by median time of sample collection date for clusters
for_plotting = lineage_sizes.iloc[:top_n, :]
for_plotting.index = clusters_meta.loc[:, 'median_days_from_01.01.20']
for_plotting = for_plotting.sort_values(by='median_days_from_01.01.20')
```

```python
# splitting by continent

# count number of conts available with a first pass
clusters_meta.index = np.arange(clusters_meta.shape[0])
all_conts = []
for i in range(for_plotting.shape[0]):
    all_conts.append(clusters_meta.loc[i, 'top_cont'].split(','))
all_conts = [x for sublist in all_conts for x in sublist]
uniq_conts = list(np.unique(all_conts))

# now split by conts
conts_for_plotting = []
for cont in uniq_conts:
    df = for_plotting.copy()
    sizes = []
    for i in range(for_plotting.shape[0]):
        split_conts = clusters_meta.loc[i, 'top_cont'].split(',')
        if cont in split_conts:
            idx = split_conts.index(cont)
            sizes.append(clusters_meta.loc[i, 'size_top_cont'].split(',')[idx])
        else:
            sizes.append(0)
    #if cont == 'North America':
        #print(orig_size)
        #print(sizes)
        #print(np.divide(np.array([int(x) for x in sizes]), orig_size))
    orig_size = clusters_meta.loc[:, 'cluster_size'].values
    df = df.mul(np.divide(np.array([int(x) for x in sizes]), orig_size), axis=0)
    df = df.astype(int)
    conts_for_plotting.append(df)
```

```python
# plot by continent
fig, axes = plt.subplots(6, 1, figsize=(12, 16))
#axes = [ax for sublist in axes for ax in sublist]
for i, ax in enumerate(axes):
    title = uniq_conts[i]
    ax.set_ylabel(title, size=20)
    if i != 5:
        if i != 0:
            ax = conts_for_plotting[i].plot(ax=ax, kind='bar', stacked=True, width=1.0, legend=False)
        else:
            ax = conts_for_plotting[i].plot(ax=ax, kind='bar', stacked=True, width=1.0, legend=True)
            ax.legend(bbox_to_anchor=(1.25, 1.01), loc='upper right', fontsize=16)
        ax.set_xticks([])
        ax.set_xlabel('')
    else:
        ax = conts_for_plotting[i].plot(ax=ax, kind='bar', stacked=True, width=1.0, legend=False)
        ax.set_xlabel('Median days since start of pandemic in cluster', size=20)
        ax.set_xticks(np.arange(1, 100, 4))
    #ax.set_ylabel('Number of sequences in cluster (log)', size=20)
    ax.tick_params(axis='both', which='major', labelsize=16)
    ax.set_ylim([0, 150000])
    #ax.set_yscale('log')
    #ax.set_ylim([10**3, 10**6])
#plt.tight_layout()
#plt.show()
```
